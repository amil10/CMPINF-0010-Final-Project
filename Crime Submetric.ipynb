## Crime! Oh Golly!
---
Counting the number of crimes that are recorded is a simple and comprehensive way to measure community safety. These thorough reports provide important insights into the security environment by detailing the crimes that have occurred in the Pittsburgh region. This metric is especially helpful to us in our quest to determine which neighborhood in Pittsburgh is the "best" since it gives us a clear way to evaluate local safety by looking at recorded criminal activity. Furthermore, analyzing this dataset may indicate that communities with lower crime rates use more successful preventative strategies and more effectively use their resources for cutting-edge security systems.

It's important to recognize this method's inherent limits, though. The reliance on crimes being reported to police is one major disadvantage. The likelihood that every crime committed in every neighborhood was reported and recorded is low, which could distort the crime count metric's accuracy. Furthermore, this metric ignores the distinction between less serious and more serious acts, making it impossible to distinguish between various categories of crimes. As a result, even while certain neighborhoods have higher rates of more serious crimes, they may nonetheless appear safer based only on lower incident counts.

While utilizing the crime count as a starting point gives us a good idea of where Pittsburgh's safest neighborhoods are, more investigation is required to gain a more complete picture. A more comprehensive understanding of local safety dynamics can be achieved by adding more data to this metric and qualitative evaluations, which will increase the certainty and accuracy of identifying the safest neighborhood in the city.

---
## Some Boring Code...

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing the dataset
crimeReport = pd.read_csv("e03a89dd-134a-4ee8-a2bd-62c40aeebc6f.tsv", sep="\t")

# Count how many times each neighborhood appears
neighborhoodCount =  crimeReport['INCIDENTNEIGHBORHOOD'].value_counts()

# Sort the values in ascending order
neighborhoodCount_sorted = neighborhoodCount.sort_values()

# Plot the results
plt.figure(figsize=(15, 35))
neighborhoodCount_sorted.plot(kind='barh')
plt.xlabel('Number of Incidents')
plt.ylabel('Neighborhood')
plt.title('Number of Incidents per Neighborhood')
plt.show()
```
